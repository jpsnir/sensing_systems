```python
import cv2 as cv
from pathlib import Path
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
data_path = Path("/media/jagatpreet/D/datasets/euroc/MH_01_easy/mav0/cam0/data")
filesnames = [f for f in data_path.iterdir() if f.is_file()]
filenames = sorted(filesnames)
img = cv.imread(data_path.joinpath(filesnames[0]).as_posix(), cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
rows, cols = img.shape
print(f' number of rows = {rows} - columns {cols}')
N_x = 10
N_y = 10
line_x = np.linspace(0, cols, N_x + 1)
line_y = np.linspace(0, rows, N_y + 1)
X, Y = np.meshgrid(line_x, line_y)
#print(f'mesh shape = {mesh.shape}')
fig, ax = plt.subplots(1, 1, figsize=(12,12))
ax.scatter(X[5], Y[5])
ax.set_xlim(-10, cols + 10)
ax.set_ylim(-10, rows + 10)
```
